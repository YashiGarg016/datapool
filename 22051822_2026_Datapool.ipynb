What can you tell about the sales based on the sizes ordered by customers?

```python
plt.hist(df['pizza_size'])
plt.xlabel('Pizza Size')
plt.ylabel('Frequency')
plt.title('Distribution of Pizza Sizes')
plt.show()
```

This histogram shows that the large (L) size pizza have the most sales whereas XXL size pizzas have the least sales.

What can you tell about the relation between pizza sizes and pizza catagories? Use boxplot analysis.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='pizza_size', y='pizza_category', data=df)
plt.xlabel('Pizza Size')
plt.ylabel('Pizza Category')
plt.title('Boxplot of Pizza Sizes and Categories')
plt.show()
```

```python
df['unit_price'].value_counts().plot(kind='bar', color=['green', 'yellow'])
```

According the pizza with unit price 20.75 have the most sales. The pizzas with unit price 35.95 have the least sales.

Which variables are highly correlated with each other?

```python
numerical_df = df.select_dtypes(include=['number'])
correlation_matrix = numerical_df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='cool')
plt.title('Correlation Matrix')
plt.show()
```

According to this heatmap, unit price and total price have the most correlation, whereas order ID and total price has the worst correlation.

What percentage of influence does different total prices of order have on the overall sales?

```python
plt.figure(figsize=(10, 10))
df['total_price'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Pie Chart of Total Prices')
plt.ylabel('')
plt.show()
```

As we can see from the piechart, total price of 20.75 has the highest contribution of 17.5%.
